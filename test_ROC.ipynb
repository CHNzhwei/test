```python
import pandas as pd
from sklearn.metrics import roc_curve, auc
import numpy as np
from sklearn import metrics
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv("C:\\Users\\pc\\Desktop\\转录组数据\\output_csv\\2021-11-22_GSE_K67_Train_Pred.csv")
train = data["Prediction_Train"].values
label = data["label"].values
print(len(data), len(label))
```

```python
fpr, tpr, thresholds = metrics.roc_curve(label, train, pos_label=1)
print("fpr",fpr)
print("tpr",tpr)
print("thresholds",thresholds)
auc = metrics.auc(fpr, tpr)
```

```python
plt.figure()
lw = 2
plt.plot(fpr, tpr, color='darkorange',
         lw=lw, label='ROC curve (area = %0.2f)' % auc)
plt.plot([0, 1], [0, 1], color='navy', lw = lw, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver operating characteristic example')
plt.legend(loc="lower right")
plt.show()
```

```python
type = 'Train'
seed = 103
date = '2021-11-29'
file = 'mdvgg_datanone'
for i in range(5):
    color_list = ['crimson', 'darkviolet', 'dodgerblue', 'lime', 'darkorange']
    one_hot = {0:[1,0,0], 1:[0,1,0], 2:[0,0,1]}
    data = pd.read_csv("C:\\Users\\pc\\Desktop\\转录组数据\\output_csv\\未排序\\%s_NHG_%s_%s_Pred_seed_%s_%sk.csv"%(date,file,type,seed,(i+1)))
    data_3 = data.loc[:,('Prediction_%s_0'%type,'Prediction_%s_1'%type,'Prediction_%s_2'%type)]
    label = data["label"].values
    data = np.array(data_3)
    oh = []
    for i in label:
        oh.append(one_hot[i])
    oh = np.array(oh)
    fpr = dict()
    tpr = dict()
    roc_auc = dict()
    for j in range(3):
        fpr[j], tpr[j], _ = roc_curve(oh[:, j], data[:, j])
        roc_auc[j] = auc(fpr[j], tpr[j])
    color_list = ['crimson', 'darkviolet', 'dodgerblue', 'lime', 'darkorange']
    plt.figure(figsize=(5,4))
    for i in  range(3):
        plt.plot(fpr[i], tpr[i], color=color_list[i],
                lw=lw, label='ROC curve(NHG %s) (area = %0.4f)'%(i,roc_auc[i]))
    plt.plot([0, 1], [0, 1], color='navy', lw = 2, linestyle='--')
    plt.xlim([-0.03, 1.03])
    plt.ylim([-0.03, 1.03])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('Receiver operating characteristic example')
    plt.legend(loc="lower right")
    plt.show()
```

```python
type = 'Valid'
date = '2021-12-10'
clf = 'k67'
model = 'mdmolmap'
data_type = 'molmap'
norm = 'none'
seed = '100101'
AUC = []
for i in range(5):
    data = pd.read_csv("C:\\Users\\pc\\Desktop\\转录组数据\\output_csv\\未排序\\%s_%s_%s_%s_%s_%s_Pred_seed_%s_%sk.csv"%(date, clf, model, data_type,norm,type,seed,(i+1)))
    valid = data["Prediction_%s_1"%type].values
    label = data["label"].values
    fpr, tpr, thresholds = metrics.roc_curve(label, valid, pos_label=1)
    auc = metrics.auc(fpr, tpr)
    AUC.append(auc)
    if i == 0:
        plt.figure(figsize=(10,7))
    lw = 2
    color_list = ['crimson', 'darkviolet', 'dodgerblue', 'lime', 'darkorange']
    color = color_list[i]
    plt.plot(fpr, tpr, color=color,
            lw=lw, label='ROC curve(fold %s) (area = %0.4f)' %((i+1),auc))
    if i == 4:
        plt.plot([0, 1], [0, 1], color='navy', lw = lw, linestyle='--')
        plt.xlim([-0.03, 1.03])
        plt.ylim([-0.03, 1.03])
        plt.xlabel('False Positive Rate')
        plt.ylabel('True Positive Rate')
        plt.title('%s 5-fold ROC Curve'%type)
        plt.legend(loc="lower right")
        plt.show()
print(np.mean(AUC))
```
